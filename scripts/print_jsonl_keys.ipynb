# Inspect Instances: problem_statement, hints_text, patch_context

Print top 5 instances from each codebase.

```python
import json
from pathlib import Path

REPOS = {
    "MiroThinker": "/home/yuansui/swe-factory-dev/internal-swe-bench-data/MiroMindAI__MiroThinker/instances_all_130.jsonl",
    "miroflow": "/home/yuansui/swe-factory-dev/internal-swe-bench-data/MiroMindAI__miroflow/instances_all_140.jsonl",
    "sd-torchtune": "/home/yuansui/swe-factory-dev/internal-swe-bench-data/MiroMindAI__sd-torchtune/instances_all_341.jsonl",
}

FIELDS = ["problem_statement", "hints_text", "patch_context"]
TOP_N = 5


def load_top_n(jsonl_path: str, n: int) -> list[dict]:
    instances = []
    with Path(jsonl_path).open("r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            instances.append(json.loads(line))
            if len(instances) >= n:
                break
    return instances
```

```python
for repo_name, jsonl_path in REPOS.items():
    instances = load_top_n(jsonl_path, TOP_N)
    print(f"\n{'#' * 80}")
    print(f"# {repo_name} — top {len(instances)} instances")
    print(f"{'#' * 80}")

    for inst in instances:
        iid = inst.get("instance_id", "unknown")
        print(f"\n{'=' * 80}")
        print(f"Instance: {iid}")
        print(f"{'=' * 80}")

        for field in FIELDS:
            val = inst.get(field, "")
            print(f"\n--- {field} ---")
            if isinstance(val, list):
                for item in val:
                    print(item if isinstance(item, str) else json.dumps(item, indent=2))
                    print()
            else:
                print(val if val else "(empty)")
```
